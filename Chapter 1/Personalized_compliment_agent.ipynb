# Personalized Compliment Agent Task


Steps to do this

1. Create a state for the graph
2. Create the compliment node
3. Add the node to the graph
4. Set the entry and finish point
5. Complie the graph and store it in a variable
6. Invoke the app with an input


```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    name:str
    message:str

def compliment_node(state:AgentState)->AgentState:
    """
    A simple node to compliment agent which takes the state and modify the state
    """

    state['message'] = f"Hey {state['name']}! You have done a marvelous job. Keep it up!"
    return state
```

```python
graph = StateGraph(AgentState)
```

```python
graph.add_node("complimentor", compliment_node)

graph.set_entry_point("complimentor")
graph.set_finish_point("complimentor")
```

```python
app = graph.compile()
```

```python
result = app.invoke({'name':"Thomson"})
result['message']
```
